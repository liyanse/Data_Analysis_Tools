```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Let's start by importing data into a pandas dataframe
flowers_df = sns.load_dataset("iris")
```

```python
flowers_df
```

```python
flowers_df.species.unique()
```

```python
plt.plot(flowers_df.sepal_length, flowers_df.sepal_width);
```

```python
sns.scatterplot(flowers_df.sepal_length, flowers_df.sepal_width);
```

```python
#We can color the plot points using the hue function and enlarge the size of points using s argument
sns.scatterplot(flowers_df.sepal_length, flowers_df.sepal_width, hue = flowers_df.species, s=100);
```

# Modifying seaborn's figures


```python
plt.figure(figsize =(12,16))
plt.title("Sepal Dimesions")

sns.scatterplot(flowers_df.sepal_length,
                flowers_df.sepal_width,
                hue = flowers_df.species, s=100);
```

# Plotting using Pandas DataFrames

```python
plt.title('Sepal Dimensions')
sns.scatterplot('sepal_length','sepal_width', hue = 'species',s=100, data = flowers_df);
```
