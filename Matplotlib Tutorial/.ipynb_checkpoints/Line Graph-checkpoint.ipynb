```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Let's start by plotting a Line chart and modify it to fit the X and Y axis
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
plt.plot(yield_apples)
```

```python
#If you want to get rid of the line <[<matplotlib.lines.Line2D at 0x243addb3490>], add a semi-colon at the end of the code.
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
plt.plot(yield_apples);
```

```python
#Let's customize the graph
years = [2010,2011,2012,2013,2014,2015]
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931]
plt.plot(years, yield_apples);
```

```python
#The next step would be to add labels to the graph on each axis
years = [2010,2011,2012,2013,2014,2015]

plt.plot(years, yield_apples);
plt.xlabel('Year')
plt.ylabel("Yields per Acre")
```

```python
#Let's try to respresent this with multiple data using two lines
years = range(2000,2012)
yield_orange= [0.905, 0.915, 0.924, 0.930, 0.939, 0.951, 0.955, 0.961, 0.966, 0.972,0.977,0.983]
yield_apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.935, 0.941, 0.945, 0.951,0.955,0.962]

plt.plot(years, yield_apples);
plt.plot(years, yield_orange);
```

```python
#Let's rename the graph and give it a legend and add marker poinnts to the lines for each point
plt.plot(years, yield_apples, marker ='o');
plt.plot(years, yield_orange, marker ='x');

plt.title("Crop Yields in Kenya")
plt.legend(['Apples','Orange']);
```

```python
#Let's improve the default styles using seaborn
sns.set_style("whitegrid")

plt.plot(years,yield_apples, 's-b')
plt.plot(years, yield_orange, 'o--r')

plt.xlabel('Year')
plt.ylabel("Yields per Acre")

plt.title("Crop Yields in Kenya")
plt.legend(['Apples','Orange']);
```
